```python
import random
```

```python
class TicTacToe():
    def __init__(self) -> None:
        self.board = [' ' for _ in range(10)]
        self.winner = None
    
    def choose_player(self):
        self.player_turn = random.choice(['x', 'o'])
        return(self.player_turn)

    def change_turn(self):
        if self.player_turn == 'x':
            self.player_turn = 'o'
            return 'o'
        else:
            self.player_turn = 'x'
            return 'x'

    def make_move(self):
        try:
            user_input_choice = int(input("Enter your choice (1-9): ").strip())
        except ValueError:
            print("Please enter a number between 1 and 9.")
            return False

        if not 1 <= user_input_choice <= 9:
            print("Choice must be between 1 and 9.")
            return False

        if self.board[user_input_choice] != ' ':
            print('Cell already taken. Choose another.')
            return False

        self.board[user_input_choice] = self.player_turn
        return True

    def show_board(self, show_indices: bool = True):
        """Print the board as a 3x3 grid.

        If show_indices is True, empty cells are shown as their position number
        (1-9) to help the player choose a move. Otherwise empty cells are blank.
        """
        def cell_text(i: int) -> str:
            if self.board[i] != ' ':
                return self.board[i]
            return str(i) if show_indices else ' '

        for start in (1, 4, 7):
            a = cell_text(start)
            b = cell_text(start + 1)
            c = cell_text(start + 2)
            print(f" {a} | {b} | {c} ")
            if start < 7:
                print("---+---+---")

    def check_winner(self):
        """Return 'x' or 'o' if there's a winner, 'Draw' if board full, otherwise None."""
        WIN_COMBINATIONS = [
            (1, 2, 3),
            (4, 5, 6),
            (7, 8, 9),
            (1, 4, 7),
            (2, 5, 8),
            (3, 6, 9),
            (1, 5, 9),
            (3, 5, 7),
        ]

        for a, b, c in WIN_COMBINATIONS:
            if self.board[a] == self.board[b] == self.board[c] != ' ':
                self.winner = self.board[a]
                return self.winner

        # Check for draw (no empty spaces in positions 1..9)
        if ' ' not in self.board[1:]:
            self.winner = 'Draw'
            return 'Draw'

        return None
```
